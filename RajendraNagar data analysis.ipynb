```python
x = df.loc[:, ['Coffee Shops', 'Salon', 'Buy', 'Young Female %', 'Genratio']]

Sx = x.cov().values
Sx = sp.linalg.inv(Sx)

mean = x.mean().values

print('Mean :', mean)
print('Inverce Covariance :\n ', Sx)
```

```python
def mahalanobisR(X,meanCol,IC):
    m = []
    for i in range(X.shape[0]):
        m.append(mahalanobis(X.iloc[i,:],meanCol,IC))
    return(m)

df['dist4'] = mahalanobisR(x,mean,Sx)
df['dist4']
```

```python
df.sort_values(by=['Rdist4'])
```

```python
m = df['Buy'] >= 18000
```

```python
df['Rdist4'] = df['dist4'].where(m, -df['dist4'])
```
