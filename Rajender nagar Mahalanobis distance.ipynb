```python
from scipy.spatial.distance import mahalanobis
```

```python
x = data.iloc[:,4:8]
#print(x.columns)
Sx = x.cov().values
Sx = np.linalg.inv(Sx)

mean = x.mean().values

print('Mean :', mean)
print('Inverce Covariance :\n ', Sx)
```

```python
def mahalanobisR(X,meanCol,IC):
    m = []
    for i in range(X.shape[0]):
        m.append(mahalanobis(X.iloc[i,:],meanCol,IC))
    return(m)

mR = mahalanobisR(x,mean,Sx)
mR
```

```python
data['mahaDist'] = mR
```

```python
data.sort_values(by=['dist1'])
```

```python
relativeDist = []
countp = 0
countn = 0
for i in range(len(data)):
    if (data.iloc[i]['Buy'] - 18000) < 0:
        relativeDist.append(data.iloc[i]['dist1']*(-1))
        countn+=1
    else :
        relativeDist.append(data.iloc[i]['dist1'])
        countp+=1
data['relDist'] = relativeDist
```

```python
dist = [1 if x>=18000 else -1 for x in data['Buy']]
data['rel'] = data['mahaDist']*dist
```

```python
clusters=range(1,10)
meanDistortions=[]
 
for k in clusters:
    model=KMeans(n_clusters=k)
    model.fit(wh1)
    prediction=model.predict(wh1)
    meanDistortions.append(sum(np.min(cdist(wh1, model.cluster_centers_, 'euclidean'), axis=1)) / wh1.shape[0])
 
#plt.cla()
plt.plot(clusters, meanDistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
```

```python
#K means Clustering 
def doKmeans(X, nclust):
    model = KMeans(nclust)
    model.fit(X)
    clust_labels = model.predict(X)
    cent = model.cluster_centers_
    return (clust_labels, cent)

clust_labels, cent = doKmeans(wh1, 3)
kmeans = pd.DataFrame(clust_labels)
wh1.insert((wh1.shape[1]),'kmeans',kmeans)
```

# Affinity Propagation 

```python
wh3 = data[['Male','Female','Coffee Shops','Salon', 'Buy', 
          'Rent','Young Female %','dist1', 'Genratio', 'relDist']] #Subsetting the data
ss = StandardScaler()
ss.fit_transform(wh3)
```

```python
final_model3 = AffinityPropagation()
final_model3.fit(wh3)
prediction3 = final_model3.predict(wh2)
 
#Join predicted clusters back to raw data
data["GROUP3"] = prediction3
print("Groups Assigned : \n")
data[["GROUP3","Class"]]
```

```python
def doAffinity(X):
    model = AffinityPropagation(damping = 0.5, max_iter = 250, affinity = 'euclidean')
    model.fit(X)
    clust_labels2 = model.predict(X)
    cent2 = model.cluster_centers_
    return (clust_labels2, cent2)

clust_labels2, cent2 = doAffinity(wh3)
affinity = pd.DataFrame(clust_labels2)
wh3.insert((wh3.shape[1]),'affinity',affinity)
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
scatter = ax.scatter(wh3['Genratio'], wh3['Young Female %'],
                     c=affinity[0],s=50)
ax.set_title('Affinity Clustering')
ax.set_xlabel('Female to male ratio')
ax.set_ylabel('Young Female Likelihood')
plt.colorbar(scatter)
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
scatter = ax.scatter(wh3['relDist'], wh3['Young Female %'],
                     c=affinity[0],s=50)
ax.set_title('Affinity Clustering')
ax.set_xlabel('Relative Distance')
ax.set_ylabel('Young Female Likelihood')
plt.colorbar(scatter)
```
