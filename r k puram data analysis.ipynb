# **Categorising according to age the males and females**

```python
data['age']= np.round(data['age'])
bins = [0, 35, 60, np.inf]
names = ['Youths', 'Adults', 'Senior Citizens']
data['age_cat'] = pd.cut(data['age'], bins, labels=names)
# pd.crosstab(data.age_cat, data.gen, margins=True)
data
```

```python
sns.countplot(data.age_cat)
data['age_cat'].value_counts()
```

```python
counts = data.groupby(['gen', 'age_cat']).age.count().unstack()
counts.plot(kind='bar', stacked=False)
plt.xlabel('Gender')
plt.ylabel('Counts')
plt.title('Gender based analysis')
```
